# 1.2 Check Stationarity

```python
# plot the data and mean to see if it looks stationary
plt.figure(figsize=(15,10))
daily_test['close'].plot(title= 'Plot to visualise and see if there is stationarity')
plt.axhline(y=daily_test['close'].mean());
```

#### Observation

+ It doesnt look stationary although its also not showing a clear trend over the time period 

+ I will now use the Dickey Fuller Test to check

```python
# Code written by Joseph Nelson.

def interpret_dftest(dftest):
    dfoutput = pd.Series(dftest[0:2], index=['Test Statistic','p-value'])
    return dfoutput
```

```python
interpret_dftest(adfuller(daily_test['close']))
```

### Observation

+ P-value shows that it is not under our 0.05 threshold so we cant reject the null hypothesis that the data is not stationary.

+ Therefore we need to difference

```python
# This loops through different values of d, when it meets the 0.05 threshold
# thenull hypothesis can be rejected and the data can be considered stationary.

# Iterate through all possible differences.
for d in range(1, 10):
    
    # Keep track of the result of each difference
    print(f'{d}: p-value = {interpret_dftest(adfuller(daily_test["close"].diff(d).dropna()))["p-value"]}')
          
    # Print succesful 
    if interpret_dftest(adfuller(daily_test['close'].diff(d).dropna()))['p-value'] < 0.05:
        print(f'Need to difference the data by d={d} then its stationary')
        break
    
    print()
```

### Obsevation
+ Need to a difference or 'D' of one to become a stationary dataset

## 1.3 Auto Regression trends

### 1.3.1 ACF

#### AR or Autoregression

+ This is where we see how much correlation the data has with itself.
+ The AR part of the model is used for looking at long term trends
+ I'll look at the Autocorrelation plot to see look for any significant correlation

```python
# Check a 30 lag time frame initially

# Set figure/axes.
fig, ax = plt.subplots(figsize=(9,5))

# Generate plot.
plot_acf(daily_test["close"], lags=30, ax=ax);
```

```python
# I will now use a longer lag time to see when the correlation starts to decline

# Set figure/axes.
fig, ax = plt.subplots(figsize=(9,5))

# Generate plot.
plot_acf(daily_test["close"], lags=730, ax=ax);
```

#### Observation

+ The data is highly correlated
+ Only after about 1 year does the data become uncorrelated.
+ There is a slight wave which is hinting at seasonality

```python
# Add a diff of one to see the impact

# Set figure/axes.
fig, ax = plt.subplots(figsize=(9,5))

# Generate plot.
plot_acf(daily_test["close"].diff(1).dropna(), lags=30, ax=ax);
```

#### Observation

+ A diff on 1 affects the ACF curve greatly. It suggests a 'P' of 1 would be appropriate. However I will check the PACF as well.

---

### 1.3.2 PACF

#### Partial Autocorrelation Plot

+ I'll look at the Partial Autocorrelation plot
+ This takes into account the correlation already been accounted for in the previous time step

```python
# 1 day  Autocorr and Partial Autocorr
fig = plt.figure(figsize=(16,16))
ax2 = fig.add_subplot(212)
fig = plot_pacf(daily_test['close'], lags=120, ax=ax2);
```

```python
# Set figure/axes.
fig, ax = plt.subplots(figsize=(9,5))

# Generate plot.
plot_acf(daily_test["close"].diff(1).dropna(), lags=120, ax=ax);
```

#### Observation

+ I would choose to have 'P' as 1 in my model as there is sharp drop off and the next lag has a positive value

+ This is explained further in this [ACF and PACF breakdown article.](http://people.duke.edu/~rnau/411arim3.htm)
